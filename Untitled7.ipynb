```python
#รวมไฟล์
import pandas as pd
import glob
import os

# 1. ระบุ Path ของโฟลเดอร์ที่ถูกต้อง (เปลี่ยน 'ชื่อโฟลเดอร์ของคุณ' ให้ตรงจริง)
# แก้ไข path ให้ชี้ไปยังโฟลเดอร์ที่เก็บไฟล์ CSV โดยตรง
path = '/content/drive/MyDrive/Data midterm เทอม 2/Input - (Operation Daily rain from API_ 2025)-20260122T061602Z-3-001/Input - (Operation Daily rain from API_ 2025)'

# 2. ดึงรายชื่อไฟล์ทั้งหมดที่ลงท้ายด้วย .csv
all_files = glob.glob(os.path.join(path, "*.csv"))

# 3. วนลูปอ่านทุกไฟล์แล้วเก็บไว้ใน List
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    # เพิ่มคอลัมน์เพื่อบอกว่ามาจากไฟล์ไหน (เผื่อไว้ตรวจสอบ)
    df['from_file'] = os.path.basename(filename)
    li.append(df)

# 4. รวมทุกไฟล์เป็น DataFrame เดียว
# ตรวจสอบก่อนว่ามีไฟล์ที่อ่านได้หรือไม่ ก่อนทำการ concat
if li:
    final_df = pd.concat(li, axis=0, ignore_index=True)
    print(f"รวมไฟล์เสร็จสิ้น! ทั้งหมด {len(all_files)} ไฟล์")
    display(final_df.head())
else:
    print(f"ไม่พบไฟล์ CSV ในโฟลเดอร์ที่ระบุ: {path}")
    print("โปรดตรวจสอบ Path หรือชื่อไฟล์ที่ถูกต้อง")
```

จากการเปรียบเทียบรายชื่อสถานีที่ควรมีตามไฟล์
Stations in the model_Hill.csv กับข้อมูลที่ปรากฏจริง
พบว่าทุกสถานีมีข้อมูลครบถ้วน จึงไม่พบ record ที่หายไป

```python
import pandas as pd

# นับจำนวน record ต่อสถานี
station_count = (
    all_data
    .groupby('Node') # Changed 'station_code' to 'Node'
    .size()
    .reset_index(name='actual_records')
)

# expected record = จำนวน record สูงสุดที่พบ
expected_records = station_count['actual_records'].max()

# คำนวณ missing และสัดส่วน
station_count['missing_records'] = expected_records - station_count['actual_records']
station_count['missing_ratio'] = station_count['missing_records'] / expected_records

# แยกสถานีที่ต้องคัดออก (>= 20%)
removed_stations = station_count[station_count['missing_ratio'] >= 0.20]

# สถานีที่เหลือ
remaining_stations = station_count[station_count['missing_ratio'] < 0.20]

# ===== แสดงผล =====
print("📌 สถานีที่ถูกคัดออก (ข้อมูลสูญหาย ≥ 20%)")
display(
    removed_stations[['Node', 'actual_records', 'missing_records', 'missing_ratio']] # Changed 'station_code' to 'Node'
)

print(f"❌ จำนวนสถานีที่ถูกคัดออก: {removed_stations.shape[0]}")
print(f"✅ จำนวนสถานีที่เหลือ: {remaining_stations.shape[0]}")
```

ถ้ามีข้อมูลหายจึงเติม median

```python
valid_stations = remaining_stations['Node']

clean_data = all_data[
    all_data['Node'].isin(valid_stations)
].copy()
```

```python
import pandas as pd

# Define the station identifier column to use consistently
station_identifier_col = 'tele_station_oldcode'

# แปลง datetime และสร้าง Day of Year
all_data['rainfall_datetime_calc'] = pd.to_datetime(
    all_data['rainfall_datetime_calc'], errors='coerce'
)
all_data['doy'] = all_data['rainfall_datetime_calc'].dt.dayofyear

# ใช้เฉพาะสถานีที่ผ่านเกณฑ์
valid_stations = remaining_stations[station_identifier_col]
data_valid = all_data[all_data[station_identifier_col].isin(valid_stations)].copy()

# สร้างช่วงวันที่ที่ควรมีข้อมูล
full_dates = pd.date_range(
    start=data_valid['rainfall_datetime_calc'].min(),
    end=data_valid['rainfall_datetime_calc'].max(),
    freq='D'
)

# expected records (station × date)
expected = pd.MultiIndex.from_product(
    [valid_stations, full_dates],
    names=[station_identifier_col, 'rainfall_datetime_calc']
).to_frame(index=False)

# รวมกับข้อมูลจริง (record ที่หายจะเป็น NaN)
merged = expected.merge(
    data_valid,
    on=[station_identifier_col, 'rainfall_datetime_calc'],
    how='left'
)

# สร้าง DOY สำหรับ record ที่เติมเข้ามา
merged['doy'] = merged['rainfall_datetime_calc'].dt.dayofyear

# คำนวณ median ตาม (station, DOY) จาก historical
median_table = (
    data_valid
    .groupby([station_identifier_col, 'doy'])['rainfall_value']
    .median()
    .reset_index()
)

# เติมค่าที่หายด้วย median
merged = merged.merge(
    median_table,
    on=[station_identifier_col, 'doy'],
    how='left',
    suffixes=('', '_median')
)

merged['rainfall_value'] = merged['rainfall_value'].fillna(
    merged['rainfall_value_median']
)

merged.drop(columns='rainfall_value_median', inplace=True)

# ===== ตรวจสอบว่าไม่มี record ไหนหายอีก =====
records_per_station = merged.groupby(station_identifier_col).size()
missing_after = merged['rainfall_value'].isna().sum()

print("จำนวน record ต่อสถานี:")
display(records_per_station)

print(f"จำนวน record ที่ยังหายหลังเติมข้อมูล: {missing_after}")
```

ใช้ median แทน mean เพราะลักษณะข้อมูลปริมาณฝนมี การกระจายไม่สมมาตรและมีค่าผิดปกติ (outliers) สูง เช่น วันฝนตกหนักมากผิดปกติ หากใช้ mean ค่าที่ได้จะถูกดึงสูงหรือต่ำเกินจริง ทำให้ค่าที่นำไปเติมบิดเบือนลักษณะจริงของข้อมูล ในขณะที่ median มีความทนทานต่อ outliers มากกว่า จึงสะท้อนค่ากลางของปริมาณฝนใน Day of Year นั้นได้เหมาะสมและเสถียรกว่าสำหรับการเติมข้อมูลจาก historical แยกตามสถานี

คำนวนปริมาณน้ำฝนรายเดือน จัดรูปแบบให้ตรงกับ output_monthly_rainfall.csv

```python
# Convert 'rainfall_datetime_calc' to datetime objects
#ต้องใช้ 'rainfall_datetime_calc' เพราะจะได้คำนวณวันเวลาได้ถูกต้อง
all_data['rainfall_datetime_calc'] = pd.to_datetime(all_data['rainfall_datetime_calc'])

# Extract year and month
all_data['year'] = all_data['rainfall_datetime_calc'].dt.year
all_data['month'] = all_data['rainfall_datetime_calc'].dt.month

# Group by station, year, and month, then sum 'rainfall_value'
monthly_rain = all_data.groupby(['tele_station_oldcode', 'year', 'month'])['rainfall_value'].sum().reset_index()

# Rename columns to match Y_monthly_rain_HII_operation
monthly_rain = monthly_rain.rename(columns={
    'tele_station_oldcode': 'Node',
    'rainfall_value': 'rain'
})

# Reorder columns to match Y_monthly_rain_HII_operation
monthly_rain = monthly_rain[['Node', 'rain', 'month', 'year']]

print("Monthly rainfall data formatted as requested:")
print(monthly_rain.head())
print("\nInfo of the new DataFrame:")
monthly_rain.info()
```

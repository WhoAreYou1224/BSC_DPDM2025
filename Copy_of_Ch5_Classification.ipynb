```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
x,y = digits.data,digits.target
```

```python
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[0])
plt.show()
```

#Model  Construction

-import

-define (hyper parameter)

-train (fitting)

-evaluate (test vs train)

```python
X_train = x[:1700]
y_train = y[:1700]
X_test = x[1700:]
y_test = y[1700:]
```

#import

```python
from sklearn.tree import DecisionTreeClassifier
```

#Difine

```python
tree0 = DecisionTreeClassifier(random_state=6)
```

#Trainning

```python
tree0.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plot_tree(tree0, feature_names=digits.feature_names, filled=True, rounded=True)
plt.show()
```

# evaluate

```python
y_predict = tree0.predict(X_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score as accuray_score
```

```python
accuray_score(y_test,y_predict)
```

## Pre-pruning (hyper parameter tuning)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree1 = DecisionTreeClassifier(max_depth=3, random_state=6)
tree1.fit(X_train,y_train)
```

```python
plt.figure(figsize=(100,50))
plot_tree(tree1, feature_names=digits.feature_names, filled=True, rounded=True,)
plt.show()
```

## evaluate

```python
y_predict1 = tree1.predict(X_test)
accuray_score(y_test,y_predict1)
```

#Tuning parameter

```python
tree2 = DecisionTreeClassifier(criterion="entropy",min_impurity_decrease=0,min_samples_split=2,ccp_alpha=0.00, random_state=6,)
tree2.fit(X_train,y_train)
accuray_score(y_test,tree2.predict(X_test))
```
